```python
def score_model(row):
    return (
                    row['R2_Test'] * 0.4 +  # Weight for R2 Test
                    row['R2_Train'] * 0.3 +  # Weight for R2 Train
                    (1 - row['MSE']) * 0.2 +  # Weight for MSE (lower is better)
                    row['Cross Validation Score'] * 0.1  # Weight for Cross Validation Score
                )

df.apply(score_model, axis=1)
```
